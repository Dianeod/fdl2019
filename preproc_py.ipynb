```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
tab=pd.read_csv('tabular_inputs_action.csv')
```

```python
tab.isnull().sum()
```

```python
tab.dtypes
```

```python
for cols in tab:
    if tab[cols].dtype=='int64':
        sb.distplot(tab[cols],kde = False)
        plt.show()
    else:
        if tab[cols].dtype=='float64':
            sb.distplot(tab[cols],kde = False)
            plt.show()
```

```python

for cols in tab:
    if tab[cols].dtype=='int64':
        tab.boxplot(column=[cols])        
        plt.show()
    else:
        if tab[cols].dtype=='float64':
            tab.boxplot(column=[cols])
            plt.show()
```
